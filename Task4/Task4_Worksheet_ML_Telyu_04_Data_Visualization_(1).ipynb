Beri sample kode python untuk dapat menampilkan persentase dari data bertipe kategorikal dalam bentuk pie chart, sehingga saya bisa copy paste ke google colab.

````python
def plot_pie_chart(column_name):
    # Hitung frekuensi data kategorikal
    category_data = data[column_name].value_counts()

    # Labels dan sizes untuk pie chart
    labels = category_data.index
    sizes = category_data.values

    # Plot pie chart
    plt.figure(figsize=(7, 7))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

    plt.axis('equal')  # Untuk memastikan pie chart berbentuk lingkaran
    plt.title(f'Percentage of {column_name}')
    plt.show()

# Contoh pemanggilan fungsi untuk beberapa kolom
plot_pie_chart('gender')
plot_pie_chart('race/ethnicity')
plot_pie_chart('lunch')
plot_pie_chart('parental level of education')
plot_pie_chart('test preparation course')
````

Beri kode untuk melihat distribusi nilai math, reading dan writing dg informasi dataframe sebagai berikut:

````python
def plot_score_distribution(data):
    # Membuat subplots untuk math, reading, dan writing score
    plt.figure(figsize=(12, 6))

    # Plot distribusi math score
    plt.subplot(1, 3, 1)
    plt.hist(data['math score'], bins=20, color='skyblue', edgecolor='black')
    plt.title('Distribusi Math Score')
    plt.xlabel('Math Score')
    plt.ylabel('Frekuensi')

    # Plot distribusi reading score
    plt.subplot(1, 3, 2)
    plt.hist(data['reading score'], bins=20, color='lightgreen', edgecolor='black')
    plt.title('Distribusi Reading Score')
    plt.xlabel('Reading Score')
    plt.ylabel('Frekuensi')

    # Plot distribusi writing score
    plt.subplot(1, 3, 3)
    plt.hist(data['writing score'], bins=20, color='salmon', edgecolor='black')
    plt.title('Distribusi Writing Score')
    plt.xlabel('Writing Score')
    plt.ylabel('Frekuensi')

    # Menampilkan plot
    plt.tight_layout()
    plt.show()

# Contoh pemanggilan fungsi
plot_score_distribution(data)
````

#Distribusi Nilai Matematika:Grafik ini menunjukkan bentuk distribusi normal yang simetris dengan puncak frekuensi di sekitar nilai 65-70dan sebagian besar siswa memperoleh nilai antara 60 hingga 80, dengan sedikit siswa di skor rendah (kurang dari 40) dan tinggi (mendekati 100).
#Distribusi Nilai Membaca:Distribusi nilai Membaca juga menunjukkan pola yang mirip dengan distribusi normal, di mana puncak frekuensi terletak di sekitar 65-70.Nilai berkisar sebagian besar antara 50 hingga 80, dengan sedikit siswa yang memiliki nilai rendah atau sangat tinggi, mirip dengan distribusi nilai Matematika.
#Distribusi Nilai Menulis:Nilai Menulis mengikuti pola yang mirip dengan Matematika dan Membaca, dengan puncak frekuensi di sekitar 65-70dan sebagian besar siswa mendapatkan nilai antara 50 hingga 80, dengan beberapa siswa yang mendapat nilai sangat rendah atau sangat tinggi.

````python
def plot_boxplot_scores(data):
    # Menyiapkan data untuk boxplot
    scores = [data['math score'], data['reading score'], data['writing score']]
    labels = ['Math Score', 'Reading Score', 'Writing Score']

    # Membuat boxplot
    plt.figure(figsize=(10, 6))
    plt.boxplot(scores, labels=labels, patch_artist=True,
                boxprops=dict(facecolor='lightblue', color='black'),
                whiskerprops=dict(color='black'),
                capprops=dict(color='black'),
                medianprops=dict(color='red'))

    plt.title('Boxplot of Math, Reading, and Writing Scores')
    plt.ylabel('Score')
    plt.grid(True)
    plt.show()

# Contoh pemanggilan fungsi
plot_boxplot_scores(data)
````

# Diagram boxplot ini menunjukkan distribusi nilai siswa dalam Matematika, Membaca, dan Menulis, dengan median sekitar 65-70 di semua subjek, menunjukkan bahwa sebagian besar siswa mencapai nilai yang relatif baik. Rentang interkuartil (IQR) pada ketiga subjek menunjukkan bahwa sebagian besar nilai berada antara 50 hingga 80, sementara nilai minimum dan maksimum yang bukan outlier berada di sekitar 30 hingga 100, menunjukkan variasi yang signifikan dalam pencapaian nilai siswa. Di sisi lain, terdapat beberapa outlier di semua subjek dengan nilai di bawah 20, mengindikasikan bahwa sebagian kecil siswa memperoleh nilai yang jauh lebih rendah dibandingkan rekan-rekan mereka.

Prompt:

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

````python
def plot_pass_fail_distribution(data):
    # Menambahkan kolom baru untuk menyimpan informasi pass/fail untuk masing-masing pelajaran
    data['math_pass'] = data['math score'] >= 75
    data['reading_pass'] = data['reading score'] >= 75
    data['writing_pass'] = data['writing score'] >= 75

    # Menghitung jumlah pass dan fail untuk setiap pelajaran
    math_pass_fail = data['math_pass'].value_counts()
    reading_pass_fail = data['reading_pass'].value_counts()
    writing_pass_fail = data['writing_pass'].value_counts()

    # Plot bar chart untuk masing-masing pelajaran
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    # Math pass/fail
    axes[0].bar(['Pass', 'Fail'], math_pass_fail, color=['lightgreen', 'salmon'])
    axes[0].set_title('Math Pass/Fail')
    axes[0].set_ylabel('Number of Students')

    # Reading pass/fail
    axes[1].bar(['Pass', 'Fail'], reading_pass_fail, color=['lightgreen', 'salmon'])
    axes[1].set_title('Reading Pass/Fail')

    # Writing pass/fail
    axes[2].bar(['Pass', 'Fail'], writing_pass_fail, color=['lightgreen', 'salmon'])
    axes[2].set_title('Writing Pass/Fail')

    # Tampilkan plot
    plt.tight_layout()
    plt.show()

# Contoh pemanggilan fungsi
plot_pass_fail_distribution(data)
````

Prompt:

Beri kode python yang menunjukkan visualisasi nilai math, reading, dan writing siswa, dengan melihat pengaruh dari gender, race/ethnicity, parental level of education, lunch, test preparation course-nya!

````python
def plot_categorical_influence(data):
    # Membuat plot untuk setiap variabel kategorikal terhadap nilai math, reading, dan writing
    categorical_columns = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']
    scores_columns = ['math score', 'reading score', 'writing score']

    for score in scores_columns:
        plt.figure(figsize=(20, 10))  # Ukuran figure lebih besar untuk menampung semua boxplot
        for i, category in enumerate(categorical_columns, 1):
            plt.subplot(2, 3, i)  # Membuat grid 2x3 untuk plot
            sns.boxplot(x=category, y=score, data=data)
            plt.title(f'{score.capitalize()} by {category.capitalize()}')
            plt.xticks(rotation=45)  # Memutar label x-axis agar lebih terbaca
        plt.tight_layout()
        plt.show()

# Contoh pemanggilan fungsi
plot_categorical_influence(data)
````

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

````python
def plot_average_score_correlation(data):
    # Tambahkan kolom rata-rata nilai untuk setiap siswa
    data['average_score'] = data[['math score', 'reading score', 'writing score']].mean(axis=1)

    # Variabel kategorikal yang akan dianalisis
    categorical_columns = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

    # Visualisasi rata-rata nilai terhadap kategori
    plt.figure(figsize=(20, 10))  # Ukuran figure yang besar
    for i, category in enumerate(categorical_columns, 1):
        plt.subplot(2, 3, i)
        sns.boxplot(x=category, y='average_score', data=data)
        plt.title(f'Average Score by {category.capitalize()}')
        plt.xticks(rotation=45)  # Memutar label agar lebih terbaca

    plt.tight_layout()
    plt.show()

    # Melihat korelasi rata-rata nilai dengan masing-masing kategori secara statistik
    for category in categorical_columns:
        print(f"\nRata-rata nilai berdasarkan {category}:")
        print(data.groupby(category)['average_score'].mean())

# Contoh pemanggilan fungsi
plot_average_score_correlation(data)
````
